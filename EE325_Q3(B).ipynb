```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats


X, Y = [], []
for line in open('hw1c2.txt', 'r'):
  values = [float(s) for s in line.split()]
  X.append(values[0])
  Y.append(values[1])

slope, intercept, r, p, std_err = stats.linregress(X, Y)

def myfunc(X):
   return  slope * X + intercept
  
mymodel = list(map(myfunc, X))

plt.scatter(X, Y)
plt.title("Height-Weight Analysis")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.plot(X, mymodel)
plt.show()
print("\nslope:" , slope)
print("\nintercept:", intercept)

error = []
for i in range(100):
  E = abs(Y[i] - myfunc(X[i]))
  error.append(E)
incr_order = sorted(error)
print("\nincreasing order of absolute error:" , incr_order)
plt.scatter(X, error)
plt.title("difference b/w true height y and intelligent guess Ë†y using model")
plt.xlabel("Height")
plt.ylabel("Error")
plt.show()



print("\nMean:",np.mean(error)) 
print("\nstd:" ,np.std(error))


```
